```python
## DECISION TREE
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
dataset
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)##giving columns title names
df
```

```python
df['class']=dataset.target ##added output columns
```

```python
df.head(10)
```

```python
df['class'].unique()#to see unique output values
```

```python
X=df.iloc[:,:-1]
```

```python
X
```

```python
Y=df.iloc[:,-1]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
df.shape
```

```python
X.shape
```

```python
Y.shape
```

```python
### after train split test we perform decision treee algoritm
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier()
```

```python
dt.fit(Xtrain,Ytrain)
```

```python
from sklearn.tree import plot_tree ##to visualise the  decision tree
```

```python
plot_tree(dt)
```

```python
predictions=dt.predict(Xtest)
```

```python
predictions
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Ytest,predictions)
```
